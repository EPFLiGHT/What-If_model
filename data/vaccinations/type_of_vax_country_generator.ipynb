```python
import pandas as pd
import numpy as np
import yaml
import pycountry
import pycountry_convert as pc
import matplotlib.pyplot as plt
```

```python
# https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/locations.csv
df = pd.read_csv("./owid_downloaded/locations.csv")
```

```python
values = list(df['vaccines'].values)
final_values = list(set([name.strip() for val in values for name in val.split(',')]))
for vaccine in final_values:
    df[vaccine] = df.apply(lambda row: 1 if vaccine in [val.strip() for val in row['vaccines'].split(',')] else 0, axis=1)

iso_codes = df['iso_code'].unique()
iso_codes = [iso_code for iso_code in iso_codes if len(iso_code) == 3]
df = df[df['iso_code'].isin(iso_codes)]

df.drop(columns=['vaccines', 'last_observation_date', 'source_name', 'source_website', 'location'], inplace = True)
df
```

```python
df.to_csv('./type_of_vax.csv', index=False)
```
