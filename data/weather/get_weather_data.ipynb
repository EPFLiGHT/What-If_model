```python
from weather_data_helpers import *
import pycountry
import pickle
import pandas as pd
import numpy as np
```

```python
country_list = list(pd.read_pickle('../countries/iso_to_capital.pkl').keys())
```

```python
final_df = download_weather_data()
final_df
```

```python
for c in country_list:
    country = pycountry.countries.get(alpha_2=c)
    if country:
        final_df = pd.concat([final_df, get_weather_country(country)])
        final_df['iso_code'] = np.where(final_df['iso_code'].isna(), country.alpha_3, final_df['iso_code'])
```

```python
final_df
```

```python
final_df.drop_duplicates().to_csv('./weather_data.csv')
```
