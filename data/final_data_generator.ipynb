## Mergining with vaccinations

```python
merged_daily = merged_daily.merge(type_of_vax_country, on='iso_code', how='left')
```

```python
vaccinations = vaccinations[vaccinations.iso_code.isin(list(merged_daily.iso_code.unique()))]
vaccinations['date'] = pd.to_datetime(vaccinations['date']).dt.date

# Getting first day of vaccinations for each country
mapping = {}
for iso_code in vaccinations.iso_code.unique():
    temp = vaccinations[vaccinations.iso_code == iso_code].sort_values('date')
    mapping[iso_code] = temp.date.values[0]
    
merged_daily = merged_daily.merge(vaccinations, on=['iso_code', 'date'], how='left')
merged_daily = merged_daily[~merged_daily['total_population_all'].isna()]
merged_daily
```

```python
# First vaccine ever recorded:
[v for k, v in sorted(mapping.items(), key=lambda item: item[1])][0]
```

```python
merged_daily['vaccines_available'] = np.where(merged_daily.date < datetime.strptime('2020-12-14', '%Y-%m-%d').date(), 0, 1)
```

```python
for iso_code in merged_daily.iso_code.unique():
    if iso_code in mapping.keys():
        starting_date = mapping[iso_code]
        merged_daily['total_vaccinations'] = np.where((merged_daily['iso_code']==iso_code) & (merged_daily['date'] < starting_date), 0, merged_daily['total_vaccinations'])
        merged_daily['people_fully_vaccinated'] = np.where((merged_daily['iso_code']==iso_code) & (merged_daily['date'] < starting_date), 0, merged_daily['people_fully_vaccinated'])
        merged_daily['people_vaccinated'] = np.where((merged_daily['iso_code']==iso_code) & (merged_daily['date'] < starting_date), 0, merged_daily['people_vaccinated'])
        merged_daily['daily_vaccinations'] = np.where((merged_daily['iso_code']==iso_code) & (merged_daily['date'] < starting_date), 0, merged_daily['daily_vaccinations'])
        
        # Considering percentages
        merged_daily['total_vaccinations'] = np.where((merged_daily['iso_code']==iso_code), merged_daily['total_vaccinations'] / merged_daily['total_population_all'], merged_daily['total_vaccinations'])
        merged_daily['people_fully_vaccinated'] = np.where((merged_daily['iso_code']==iso_code), merged_daily['people_fully_vaccinated'] / merged_daily['total_population_all'], merged_daily['people_fully_vaccinated'])
        merged_daily['people_vaccinated'] = np.where((merged_daily['iso_code']==iso_code), merged_daily['people_vaccinated'] / merged_daily['total_population_all'], merged_daily['people_vaccinated'])
        merged_daily['daily_vaccinations'] = np.where((merged_daily['iso_code']==iso_code), merged_daily['daily_vaccinations'] / merged_daily['total_population_all'], merged_daily['daily_vaccinations'])
    else:
        print(iso_code)
```

```python
merged_daily.sort_values(by=['iso_code', 'date']).to_csv('./final_data.csv', index=False)
```

## SwissRE platform data generation

```python
# Extracting datast on vaccines for the platform
vacc = merged_daily[['iso_code', 'date', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'daily_vaccinations']]
for iso_code in vacc.iso_code.unique():
    if iso_code in mapping.keys():
        starting_date = mapping[iso_code]
        vacc = vacc.drop(vacc[(vacc.date < starting_date) & (vacc.iso_code == iso_code)].index)
    else:
        vacc = vacc.drop(vacc[(vacc.iso_code == iso_code)].index)
vacc
```

```python
vacc.sort_values(by=['iso_code', 'date']).to_csv('./vaccination_percentages.csv', index = False)
```
