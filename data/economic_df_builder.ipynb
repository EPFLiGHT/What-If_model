```python
# Using as economic indicator unemployment rate. Interpolating the quarterly data.
unemp = data_swissre[['date', 'iso_code', 'unemployment_rate_2019_12_31', 'unemployment_rate_2020_03_31', 'unemployment_rate_2020_06_30', 'unemployment_rate_2020_09_30', 'unemployment_rate_2020_12_31', 'unemployment_rate_2021_03_31', 'unemployment_rate_2021_06_30']]
```

```python
unemp = unemp[~unemp.unemployment_rate_2020_03_31.isna()]
unemp
```

```python
eco = ['unemployment_rate_2020_03_31', 'unemployment_rate_2020_06_30', 'unemployment_rate_2020_09_30', 'unemployment_rate_2020_12_31', 'unemployment_rate_2021_03_31', 'unemployment_rate_2021_06_30']
#for i, indicator in enumerate(reversed(eco[1:])):
    #unemp[indicator] = unemp[indicator] - unemp[eco[-(i+2)]]
    
for indicator in eco:
    unemp[indicator] = unemp[indicator] - unemp['unemployment_rate_2019_12_31']
```

```python
unemp['unemployment_rate_2019_12_31'] = 0
unemp
```

```python
iso_list = unemp.iso_code.unique()
dates = list(pd.date_range('2020-03-31', '2021-06-30').strftime('%Y-%m-%d'))
dates_total = dates*len(iso_list)
iso_list_total = list(np.array([[iso_code]*len(dates) for iso_code in iso_list]).flatten())
unemployment_rate_idx= [np.nan for i in iso_list_total]
```

```python
df = pd.DataFrame({'iso_code': iso_list_total, 'date':dates_total, 'unemployment_rate_idx':unemployment_rate_idx})
```

```python
for iso_code in iso_list:
    #df['unemployment_rate_idx'] = np.where((df['iso_code']== iso_code)&(df['date'] == '2019-12-31'), unemp[unemp['iso_code'] == iso_code]['unemployment_rate_2019_12_31'].values[0], df['unemployment_rate_idx'])
    df['unemployment_rate_idx'] = np.where((df['iso_code']== iso_code)&(df['date'] == '2020-03-31'), unemp[unemp['iso_code'] == iso_code]['unemployment_rate_2020_03_31'].values[0], df['unemployment_rate_idx'])
    df['unemployment_rate_idx'] = np.where((df['iso_code']== iso_code)&(df['date'] == '2020-06-30'), unemp[unemp['iso_code'] == iso_code]['unemployment_rate_2020_06_30'].values[0], df['unemployment_rate_idx'])
    df['unemployment_rate_idx'] = np.where((df['iso_code']== iso_code)&(df['date'] == '2020-09-30'), unemp[unemp['iso_code'] == iso_code]['unemployment_rate_2020_09_30'].values[0], df['unemployment_rate_idx'])
    df['unemployment_rate_idx'] = np.where((df['iso_code']== iso_code)&(df['date'] == '2020-12-31'), unemp[unemp['iso_code'] == iso_code]['unemployment_rate_2020_12_31'].values[0], df['unemployment_rate_idx'])
    df['unemployment_rate_idx'] = np.where((df['iso_code']== iso_code)&(df['date'] == '2021-03-31'), unemp[unemp['iso_code'] == iso_code]['unemployment_rate_2021_03_31'].values[0], df['unemployment_rate_idx'])
    df['unemployment_rate_idx'] = np.where((df['iso_code']== iso_code)&(df['date'] == '2021-06-30'), unemp[unemp['iso_code'] == iso_code]['unemployment_rate_2021_06_30'].values[0], df['unemployment_rate_idx'])
```

```python
df['unemployment_rate_idx'] = df['unemployment_rate_idx'].interpolate(method='linear')
df['date'] =pd.to_datetime(df['date'])
```

```python
final = data_swissre.merge(df, on=['iso_code', 'date'])
final[final['iso_code'] == 'ITA']
```

```python
quarters = [('2020-03-31', '2020-06-30'), ('2020-06-30', '2020-09-30'), 
            ('2020-09-30', '2020-12-31'), ('2020-12-31', '2021-03-31'), 
            ('2021-03-31', '2021-06-30')]
```

```python
weights_df = pd.DataFrame({'date':dates, 'weight':[1 for i in range(len(dates))]})
weights_df['date'] = pd.to_datetime(weights_df['date']).dt.date
epsilon = 1e-4
for quarter in quarters:
    start = datetime.strptime(quarter[0], '%Y-%m-%d').date()
    end = datetime.strptime(quarter[1], '%Y-%m-%d').date()
    interval_size = (end-start).days
    denominator = interval_size/2
    for i in range(interval_size):
        w = (np.abs(denominator-i) / denominator) + epsilon
        weights_df['weight'] = np.where(weights_df['date'] == start + timedelta(days=i), w, weights_df['weight'])
        
```

```python
final['date'] = pd.to_datetime(final['date']).dt.date
final[final['iso_code'] == 'ITA']
```

```python
final = final.merge(weights_df, on='date')
final
```
