```python
import numpy as np
import pandas as pd
import pycountry_convert as pc
from confidence_interval import confidence_interval as ci
```

```python
RL_re = pd.read_csv("RL_re_complete.csv", delimiter=';') #("RL_re_february.csv") #("RL_re_december.csv") #("RL_re_complete.csv", delimiter=';')
RL_unemp = pd.read_csv("RL_unemp_complete.csv", delimiter=';') #("RL_unemp_february.csv") #("RL_unemp_complete.csv", delimiter=';')#("RL_unemp_december.csv")

data_swissre = pd.read_csv('data/final_data.csv', parse_dates=['date']).set_index('date')
data_swissre_economic = pd.read_csv('./data/economic_weighted.csv', parse_dates=['date']).set_index('date')
```

```python
iso_codes_economic = ['AUT','BEL', 'BGR', 'HRV', 'CYP', 'CZE', 'DNK', 'EST', 'FIN', 'FRA', 'DEU', 'GRC', 'HUN', 'IRL', 'ITA', 'LVA', 'LTU', 'LUX', 'MLT', 'NLD', 'PRT', 'ROU', 'SVK', 'SVN', 'ESP', 'SWE', 'GBR', 'CHE', 'NOR']
```

```python
data_swissre = data_swissre[data_swissre.iso_code.isin(iso_codes_economic)]
data_swissre_economic = data_swissre_economic[data_swissre_economic.iso_code.isin(iso_codes_economic)]

```

```python
starting_date = '2021-04-11' # '2020-12-06' #'31-01-2021' #'2021-04-11'#2020-12-06' 
ending_date = '2021-05-09' # '2021-01-03' #28-02-2021' #'2021-05-09'#2021-01-03'
```

```python
initial_re = data_swissre[data_swissre.index == starting_date][['iso_code', 'epiforecasts_effective_reproduction_number_median']].rename(columns={'epiforecasts_effective_reproduction_number_median' : 'initial_re'})
initial_re
```

```python
true_final_re = data_swissre[data_swissre.index == ending_date][['iso_code', 'epiforecasts_effective_reproduction_number_median']].rename(columns={'epiforecasts_effective_reproduction_number_median' : 'true_final_re'})
true_final_re
```

```python
RL_re = RL_re[['iso_code', 'final_re']]
RL_re
```

```python
RL_re_merged = initial_re.merge(true_final_re, on='iso_code').merge(RL_re, on='iso_code')
RL_re_merged['difference_re'] = RL_re_merged['true_final_re'] - RL_re_merged['final_re']
```

```python
ci((RL_re_merged['initial_re'] - RL_re_merged['true_final_re']).values)
```

```python
# february: -0.099
# december: 0.083
# april: 0.09

np.array([-0.099, 0.083, 0.09]).mean().round(3)
```

```python
# february: 0.194
# december: 0.255
# april: 0.301

(RL_re_merged['initial_re'] - RL_re_merged['final_re']).mean().round(3)
```

```python
np.array([0.194, 0.255, 0.301]).mean().round(3)
```

```python
RL_re_merged = RL_re_merged[['iso_code', 'initial_re', 'difference_re']].round(3)
```

## Average and std

```python
february = pd.read_excel('february_output.xlsx').iloc[2:][['Unnamed: 0', 'Unnamed: 2', 'Unnamed: 4']].rename(columns={'Unnamed: 0':'Country', 'Unnamed: 2':'Difference RE (february)', 'Unnamed: 4':'Difference UER (february)'})
december = pd.read_excel('december_output.xlsx').iloc[2:][['Unnamed: 0', 'Unnamed: 2', 'Unnamed: 4']].rename(columns={'Unnamed: 0':'Country', 'Unnamed: 2':'Difference RE (december)', 'Unnamed: 4':'Difference UER (december)'})
april = pd.read_excel('april_output.xlsx').iloc[2:][['Unnamed: 0', 'Unnamed: 2', 'Unnamed: 4']].rename(columns={'Unnamed: 0':'Country', 'Unnamed: 2':'Difference RE (april)', 'Unnamed: 4':'Difference UER (april)'})
```

```python
total = february.merge(december, on='Country').merge(april, on='Country')
total['Difference RE (avg)'] = total[['Difference RE (february)', 'Difference RE (december)', 'Difference RE (april)']].mean(axis=1)
total['Difference UER (avg)'] = total[['Difference UER (february)', 'Difference UER (december)', 'Difference UER (april)']].mean(axis=1)
total['Difference RE (std)'] = total[['Difference RE (february)', 'Difference RE (december)', 'Difference RE (april)']].std(axis=1)
total['Difference UER (std)'] = total[['Difference UER (february)', 'Difference UER (december)', 'Difference UER (april)']].std(axis=1)

total[['Country', 'Difference RE (avg)', 'Difference RE (std)', 'Difference UER (avg)', 'Difference UER (std)']]#.to_excel("rl_avg_std_output.xlsx")
```
